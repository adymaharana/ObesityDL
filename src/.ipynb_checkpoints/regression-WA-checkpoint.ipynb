```python
%matplotlib inline
from sklearn.cross_validation import KFold
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet, SGDRegressor
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing
from sklearn.metrics import r2_score
import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
import random
import csv
import seaborn as sns
```

```python
def plot_fit(y_pred, y_actual, title):
    plt.plot(y_actual, y_pred, markersize=5, linestyle='None', marker = 'o')
    plt.plot([min(y_actual),max(y_actual)],[min(y_actual),max(y_actual)],linestyle='--')
    plt.xlabel('observed')
    plt.ylabel('predicted')
    plt.title(title)
    plt.show()

def filter_dataset(X, y):
    inds = []
    mean = np.mean(y)
    std = np.std(y)
    
    #for i in range(0,X.shape[0]):
    #    if (y[i]>=1) and (y[i]<=1000):
    #        inds.append(i)
    
    for i in range(0, y.shape[0]):
        if y[i]>=(mean-(2.0*std)) and y[i]<=(mean+(2.0*std)):
            inds.append(i)
    new_features = X[np.array(inds), :]
    new_pcc = [y[ind] for ind in inds]
    return (new_features, new_pcc)

def pca_X(X, n_components=2):
    pca = PCA(n_components)
    scaler = StandardScaler(with_mean=True, with_std=False)
    X_train = scaler.fit_transform(pca.fit_transform(X))
    print("Explained Variance: " + str(sum(pca.explained_variance_ratio_)))
    return (X_train, scaler, pca)

def lin_reg(X_train, y):
    # Create linear regression object
    linreg = LinearRegression()

    # Train the model using the training sets
    linreg.fit(X_train,y)
    return linreg

def error_analysis(clf, X, y):
    # Compute RMSE on training data
    # p = np.array([linreg.predict(xi) for xi in x])
    p = clf.predict(X)
    # Now we can constuct a vector of errors
    err = abs(p-y)
    
    total_error = np.dot(err,err)
    # Compute RMSE
    rmse = np.sqrt(total_error/len(p))
    return (p, rmse)
    
def ridge_reg(X_train, y):
    
    print('Ridge Regression')
    print('alpha\t RMSE_train\t RMSE_10cv\n')
    alpha = np.linspace(.01,20,10)
    t_rmse = np.array([])
    cv_rmse = np.array([])

    for a in alpha:
        ridge = Ridge(fit_intercept=True, alpha=a)

        # computing the RMSE on training data
        ridge.fit(X_train,y)
        p = ridge.predict(X_train)
        err = p-y
        total_error = np.dot(err,err)
        rmse_train = np.sqrt(total_error/len(p))

        # computing RMSE using 10-fold cross validation
        kf = KFold(len(X_train), n_folds=5)
        xval_err = 0
        for train, test in kf:
            ridge.fit(X_train[train], y[train])
            p = ridge.predict(X_train[test])
            err = p - y[test]
            xval_err += np.dot(err,err)
        rmse_10cv = np.sqrt(xval_err/len(X_train))

        t_rmse = np.append(t_rmse, [rmse_train])
        cv_rmse = np.append(cv_rmse, [rmse_10cv])
        print('{:.3f}\t {:.4f}\t\t {:.4f}'.format(a,rmse_train,rmse_10cv))
        
def sv_reg(X_train, y, kernel='rbf', degree=2):
    if kernel == 'rbf':
        svr = SVR(kernel='rbf', C=10, gamma=0.05)
    elif kernel == 'linear':
        svr = SVR(kernel='linear', C=100, verbose = True)
    elif kernel == 'poly':
        svr = SVR(kernel='poly', C=10, degree=degree)
    
    clf = svr.fit(X_train,y)
    return clf

def rf_reg(X_train, y):
    model = RandomForestRegressor(n_estimators=100, max_features=10)
    model.fit(X_train, y)
    return model

def cross_val(X, y, clf, n=5, pca_components=50):
    
    X, y = filter_dataset(X, y)

    index_shuf = list(range(len(y)))
    random.shuffle(index_shuf)

    #X = X[np.array(index_shuf), :]
    #y = [y[ind] for ind in index_shuf]
    
    y_pred = []
    y_actual = []
    rmse_train = 0
    rmse_test = 0
    r2_sum = 0
    datasize = X.shape[0]
    for i in range(0,n):
        print('--------- Fold ' + str(i+1) + ' in ' + str(n) + '-fold Cross-Validation ----------')
    
        trainidx = list(range(0,int(i*datasize/n))) + list(range(int((i+1)*datasize/n),datasize))
        testidx = list(range(int(i*datasize/n),int((i+1)*datasize/n)))

        #scaler = StandardScaler().fit(X)
        #X = scaler.transform(X)

        X_train, scaler, pca = pca_X(X[trainidx,:], pca_components)
        y_train = np.array([y[idx] for idx in trainidx])
        X_train = np.array([np.concatenate((v,[1])) for v in X_train])
        
        clf.fit(X_train,y_train)
        p_train, rmse_train_split = error_analysis(clf, X_train, y_train)
        X_test = scaler.transform(pca.transform(X[testidx,:]))
        X_test = np.array([np.concatenate((v,[1])) for v in X_test])
        y_test = np.array([y[idx] for idx in testidx])
        p_test, rmse_test_split = error_analysis(clf, X_test, y_test)
        
        y_pred.extend(p_test.tolist())
        y_actual.extend(y_test.tolist())
        rmse_train += rmse_train_split
        rmse_test += rmse_test_split
        r2_sum += clf.score(X_test, y_test)
        
        print('RMSE_Train, RMSE_Test: ' + str(rmse_train_split) + ', ' + str(rmse_test_split))
        print('R2 Score for Train, Test: ' + str(clf.score(X_train, y_train)) + ', ' + str(clf.score(X_test, y_test)))
        print('')
    
    rmse_train = rmse_train/n
    rmse_test = rmse_test/n
    r2 = r2_score(y_actual, y_pred)
    r2_new = r2_sum/n
    
    X, _, _ = pca_X(X, pca_components)
    y = np.array(y)
    X = np.array([np.concatenate((v,[1])) for v in X])
    clf.fit(X, y)
    
    print('')
    print('--------- Results from ' + str(n) + '-fold Cross-Validation ----------')
    print('Size of Dataset: ' + str(X.shape[0]))
    print('Average RMSE for Train Sets: ' + str(rmse_train))
    print('Average RMSE for Test Sets: ' + str(rmse_test))
    print('Cross-Validated R2 Score: ' + str(r2))
    print('Cross-Validated R2 Score (new method): ' + str(r2_new))
    print('R2 Score Fit for Entire Dataset: ' + str(clf.score(X,y)))
    
    return y_pred, y_actual
```

```python
import json
def poi_features(poijsonfile):
    with open(poijsonfile, 'r') as f:
        poi_list = json.load(f)
    print(len(poi_list))
    
    poi_keys = []
    for poi_dict in poi_list:
        keys = list(poi_dict.keys())
        for key in keys:
            if key not in poi_keys:
                poi_keys.append(key)
    poi_keys.remove('filename')
    poi_keys.remove('loc')
    poi_keys.remove('point_of_interest')
    poi_keys.remove('establishment')
    poi_keys.remove('locality')
    poi_keys.remove('political')
    
    print(len(poi_keys))
    print(poi_keys)
    poi_feats = []
    for poi_dict in poi_list:
        temp = [0]*len(poi_keys)
        for i in range(0, len(poi_keys)):
            try:
                temp[i] = poi_dict[poi_keys[i]]
            except KeyError:
                continue
        poi_feats.append(temp)
        
    return np.array(poi_feats)
```
